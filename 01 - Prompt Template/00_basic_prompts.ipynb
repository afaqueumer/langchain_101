## Dynamic Prompt

```python
from langchain_core.prompts import PromptTemplate
```

```python
## single variable

template = PromptTemplate.from_template(template='Greet this person in 5 languages. The name of the person is {name}')
template
```

```python
template.format(name="Optimus")
```

```python
## Multiple 

template = PromptTemplate.from_template(template= "Greet {name} in {num_languages} languages. The person is from {country}."
)

prompt = template.format(name="Optimus", num_languages=5, country="Cybertron")
print(prompt)
```

```python
## By Define Input Variables

template_str = (
    "Write a {tone} greeting for {name} in {num_languages} languages."
)

template = PromptTemplate(
    template=template_str,
    input_variables=["tone", "name", "num_languages"],
    
)

prompt = template.format(tone="formal", name="Optimus", num_languages=3)
print(prompt)
```

```python
## By dictionary unpacking

template = PromptTemplate.from_template(
    template= "Write a {tone} greeting for {name} in {num_languages} languages."
)

params = {"name": "Optimus", "num_languages": 5, "tone": "funny"}

prompt = template.format(**params)
print(prompt)
```
